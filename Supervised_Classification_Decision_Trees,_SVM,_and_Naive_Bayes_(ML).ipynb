3.Question 3:What is Pre-Pruning in Decision Trees?


Pre-Pruning (also called Early Stopping) is a technique used in Decision Trees to stop the tree from growing too deep during the training process.

The goal of pre-pruning is to prevent overfitting by limiting the complexity of the tree.

Why Pre-Pruning is Used

If a decision tree grows too large:

It may fit the training data very well

But it may perform poorly on new/unseen data

Pre-pruning helps by stopping further splits when they do not significantly improve the model.

Common Pre-Pruning Conditions

A decision tree may stop splitting when:

Maximum depth is reached
Example: max_depth = 5

Minimum number of samples in a node
Example: min_samples_split = 10

Minimum samples required in a leaf node
Example: min_samples_leaf = 5

Information gain is too small
If splitting does not improve the model significantly.

4.Question 4:Write a Python program to train a Decision Tree Classifier using Gini
Impurity as the criterion and print the feature importances (practical).
Hint: Use criterion='gini' in DecisionTreeClassifier and access .feature_importances_.

```python

# Import required libraries
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# Sample dataset
data = {
    'Age': [25, 30, 45, 35, 22],
    'Income': [50000, 60000, 80000, 65000, 45000],
    'Student': [1, 0, 0, 1, 1],
    'Credit_Rating': [0, 1, 0, 1, 0],
    'Buy_Product': [1, 0, 1, 1, 0]
}

df = pd.DataFrame(data)

# Features (X) and Target (y)
X = df[['Age', 'Income', 'Student', 'Credit_Rating']]
y = df['Buy_Product']

# Create Decision Tree model using Gini Impurity
model = DecisionTreeClassifier(criterion='gini')

# Train the model
model.fit(X, y)

# Print feature importances
print("Feature Importances:")
for feature, importance in zip(X.columns, model.feature_importances_):
    print(feature, ":", importance)



```

Explanation

criterion='gini' → Uses Gini Impurity to decide splits in the decision tree.

model.fit(X, y) → Trains the decision tree model.

feature_importances_ → Shows how important each feature is in making predictions.

Question 5: What is a Support Vector Machine (SVM)?


Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression problems, but it is mainly used for classification tasks.

SVM works by finding the best boundary (called a hyperplane) that separates different classes in the dataset.

Key Idea of SVM

The goal of SVM is to find the optimal hyperplane that maximizes the margin between two classes.

Hyperplane → The decision boundary that separates classes.

Margin → The distance between the hyperplane and the nearest data points.

Support Vectors → The data points closest to the hyperplane that help define the boundary.

Question 6: What is the Kernel Trick in SVM?


Ans:The Kernel Trick in Support Vector Machine (SVM) is a technique used to transform non-linear data into a higher-dimensional space so that it becomes linearly separable.

In simple terms, if the data cannot be separated by a straight line, the kernel trick helps map the data into a higher dimension where a linear boundary can separate it.

Why Kernel Trick is Needed

Sometimes datasets look like this:

Points from two classes are mixed in a circular or complex pattern

A straight line cannot separate them

By using the kernel trick, SVM projects the data into a higher dimension, where it becomes easier to separate using a linear hyperplane.

Question 7: Write a Python program to train two SVM classifiers with Linear and RBF
kernels on the Wine dataset, then compare their accuracies.
Hint:Use SVC(kernel='linear') and SVC(kernel='rbf'), then compare accuracy scores after fitting
on the same dataset.
(Include your Python code and output in the code box below.)

Ans:

```python

# Import required libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load Wine dataset
wine = datasets.load_wine()
X = wine.data
y = wine.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train SVM with Linear Kernel
svm_linear = SVC(kernel='linear')
svm_linear.fit(X_train, y_train)

# Predict using Linear Kernel model
y_pred_linear = svm_linear.predict(X_test)

# Train SVM with RBF Kernel
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train, y_train)

# Predict using RBF Kernel model
y_pred_rbf = svm_rbf.predict(X_test)


```

Explanation

SVC(kernel='linear') → Trains SVM using a linear decision boundary.

SVC(kernel='rbf') → Trains SVM using a non-linear RBF kernel.

train_test_split() → Splits data into training and testing sets.

accuracy_score() → Measures how well the model predicts the correct class.

Interpretation

The model with the higher accuracy performs better on the dataset.

Sometimes Linear kernel performs better, while for complex patterns RBF kernel performs better.

Question 10: Breast Cancer Dataset
Write a Python program to train a Gaussian Naïve Bayes classifier on the Breast Cancer
dataset and evaluate accuracy.
Hint:Use GaussianNB() from sklearn.naive_bayes and the Breast Cancer dataset from
sklearn.datasets.
(Include your Python code and output in the code box below.)

Ans:

```python

# Import required libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Create Gaussian Naive Bayes model
model = GaussianNB()

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print accuracy
print("Accuracy of Gaussian Naive Bayes model:", accuracy)
```
